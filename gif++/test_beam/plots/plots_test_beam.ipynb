```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib as mpl
import numpy as np

import ClusteRateWP as cr

import mplhep as hep
plt.style.use(hep.style.CMS)
```

```python
def plot_current_hv_eff(path ,file, legend, gap):
    
    # List to pass the current for each csv
    current_list = []
    current_error_list = []
    # List to pass the cluster size for each csv
    hv_list = []
        
    for f in file:
        # Read the csv file as a panda dataframe
        df = pd.read_csv(path + f)       
            
        # Takes the current for each gap     
        imon_gap = df['imon_' + gap]
        # Takes the current error
        imon_gap_error = df['imon_err_' + gap]
        
        # Takes the eff voltage
        eff_volt = df['hveff_' + gap]
       
        # Insert the desired values on the lists
        current_list.append(imon_gap)
        current_error_list.append(imon_gap_error)
        hv_list.append(eff_volt)
        
    # Figure and axis
    fig, ax = plt.subplots()
    # Attributes xaxis and yaxis
    
    colors = np.array(['#1f77b4', '#aec7e8', '#ff7f0e', '#ffbb78', '#2ca02c',
                       '#98df8a', '#d62728', '#ff9896', '#9467bd', '#c5b0d5',
                       '#8c564b', '#c49c94', '#e377c2', '#f7b6d2', '#7f7f7f',
                       '#c7c7c7', '#bcbd22',])
    markers = np.array(["1", "8", "s", "p", "*", "+", "X", "d", "P", ">", "x", "D", "H"])
    
    # Loop used for plot the data for each csv file
    for (ef, er, vo, leg, col, ma) in zip(current_list, current_error_list, hv_list, legend, colors, markers):
        #ax.plot(vo, ef, marker='s', linestyle=None, linewidth=0)
        plt.errorbar(vo, ef, yerr=er, marker=ma, mfc=col, markersize=10, linestyle='', label=leg), #marker='.', linestyle=None, linewidth=0)
        ax.legend(loc='best', fontsize='17')

    # Xlabel
    ax.xaxis.set_label_coords(0.5, -0.055)
    ax.set_xlabel(r'$HV_{eff} (V)$', fontsize = 22)

    # Ylabel
    ax.yaxis.set_label_coords(-0.09, 0.86)
    ax.set_ylabel(r'Current ($\mu A$)', fontsize = 22)
    #plt.legend('labels')

    # CMS format
    hfont = {'fontname':'Helvetica'}    
    plt.text(0.12, 0.89, "CMS", fontdict=hfont,  fontweight='bold', transform=plt.gcf().transFigure) # Value for on top: 0.17, 0.89, inside plot: 0.17, 0.80
    plt.text(0.22, 0.89, "Preliminary", fontdict=hfont, style='italic',fontsize = 22, transform=plt.gcf().transFigure) # Value for on top: 0.27, 0.89, inside plot: 0.27, 0.80

    ### If ones would like to move the scientific notation
    #t = ax.yaxis.get_offset_text()
    #t.set_x(0.008)

    plt.grid()
    
    plt.savefig('Current_VS_HVeff_' + gap + '.png')
```

```python
""" This function is used to plot the efficiency in function of HVeff 

The argument file is a list with csv files for each ABS (rates) with the following columns: 

For each csv file one should include a legend for the rates/abs.

How you call the function:

file = ['data_RE1_1_004662.csv']  -> List with the csv files containing the data
legend = [r'$xx Hz/cm^2$']        -> Legend for each csv file
chamber_hv = 'hveff_RE1_1_001-TN' -> Name of the column containing the voltages (inside the csv file). Normally the name is hveff_ + chamber + gap

plot_eff(file, legend)

"""
def sigmoid(x, a, b, c):
    return a/(1 + np.exp(-b*(x-c)))

def plot_eff(path, file, legend, chamber_hv, fit):
        
    # List to pass the efficiency for each csv
    eff_list = []
    # List to pass the efficiency error for each csv
    eff_error_list = []
    voltage_list = []
    
    for f in file:
        # Read the csv file as a panda dataframe
        df = pd.read_csv(path + f)
        # Takes the effective voltage
        try: 
            voltage = df['hveff_' + chamber_hv[0]].astype(float)
        except:
            voltage = df['hveff_' + chamber_hv[1]].astype(float)
            
        # Takes the efficiency     
        efficiency = df['efficiencyMuon']
        efficiency_error = df['efficiencyMuon_err']
        # Insert the desired values on the lists
        eff_list.append(efficiency)
        eff_error_list.append(efficiency_error)
        voltage_list.append(voltage)
        
    # Figure and axis
    fig, ax = plt.subplots()
    # Attributes xaxis and yaxis
    
    colors = np.array(['#1f77b4', '#aec7e8', '#ff7f0e', '#ffbb78', '#2ca02c',
                       '#98df8a', '#d62728', '#ff9896', '#9467bd', '#c5b0d5',
                       '#8c564b', '#c49c94', '#e377c2', '#f7b6d2', '#7f7f7f',
                       '#c7c7c7', '#bcbd22',])
    markers = np.array(["1", "8", "s", "p", "*", "+", "X", "d", "P", ">", "x", "D", "H"])
    
    # Loop used for plot the data for each csv file
    for (ef, er, vo, leg, col, ma) in zip(eff_list, eff_error_list, voltage_list, legend, colors, markers):
        #ax.plot(vo, ef, marker='s', linestyle=None, linewidth=0)
        plt.errorbar(vo, ef, yerr=er, marker=ma, markersize=10, linestyle='', label=leg, mfc=col), #marker='.', linestyle=None, linewidth=0)
        ax.legend(loc='best', fontsize=15)
        
        if fit:
            from scipy import optimize
        
            params, params_covariance = optimize.curve_fit(sigmoid, vo, ef, p0=[96.6, 0.007, 7093])
            plt.plot(vo, sigmoid(vo, params[0], params[1], params[2]), color=col)

    # Xlabel
    ax.xaxis.set_label_coords(0.5, -0.055)
    ax.set_xlabel(r'$HV_{eff} (V)$', fontsize = 22)

    # Ylabel
    ax.yaxis.set_label_coords(-0.09, 0.86)
    ax.set_ylabel(r'Efficiency', fontsize = 22)
    #plt.legend('labels')

    # CMS format
    hfont = {'fontname':'Helvetica'}    
    plt.text(0.12, 0.89, "CMS", fontdict=hfont,  fontweight='bold', transform=plt.gcf().transFigure) # Value for on top: 0.17, 0.89, inside plot: 0.17, 0.80
    plt.text(0.22, 0.89, "Preliminary", fontdict=hfont, style='italic',fontsize = 22, transform=plt.gcf().transFigure) # Value for on top: 0.27, 0.89, inside plot: 0.27, 0.80


    ### If ones would like to move the scientific notation
    #t = ax.yaxis.get_offset_text()
    #t.set_x(0.008)

    plt.grid()
    
    plt.savefig('Efficiency_VS_HVeff_' + chamber_hv[0] + '.png')
```

```python
def efficiency_gamma_rate_WP(path, gas):
    
    # Figure and axis
    fig, ax = plt.subplots()
    
    for f in gas.keys():
        leg = f
        # List to pass the efficiency for each csv
        eff_list = []
        # List to pass the efficiency error for each csv
        eff_error_list = []
        gamma_rate_list = []
        for c in gas[f]:
            # Read the csv file as a panda dataframe
            df = pd.read_csv(path + f + '/' + c)

            # Takes the gamma cluster size    
            gamma_cs = df['gamma_CLS_WP_muon_corr'][0]
            # Takes the hit rate
            gamma_hit_rate = df['noiseGammaRate_WP_muon_corr'][0]
            # Compute the cluster rate
            cluster_rate = gamma_hit_rate / gamma_cs
            gamma_rate = cluster_rate / df['eff_WP_muon_corr'][0] * 100

            # Takes the efficiency     
            efficiency = df['eff_WP_muon_corr'][0]
            #efficiency_error = df['efficiencyMuon_err']
            # Insert the desired values on the lists
            eff_list.append(efficiency)
            eff_error_list.append(0)
            gamma_rate_list.append(gamma_rate)
        
        colors = np.array(['#1f77b4', '#aec7e8', '#ff7f0e', '#ffbb78', '#2ca02c',
                       '#98df8a', '#d62728', '#ff9896', '#9467bd', '#c5b0d5',
                       '#8c564b', '#c49c94', '#e377c2', '#f7b6d2', '#7f7f7f',
                       '#c7c7c7', '#bcbd22',])
        markers = np.array(["1", "8", "s", "p", "*", "+", "X", "d", "P", ">", "x", "D", "H"])
        
        # Attributes xaxis and yaxis
        plt.errorbar(gamma_rate_list, eff_list, yerr=eff_error_list, marker='D', markersize='10',
                     linestyle='', label=leg)
        ax.legend(loc='best', fontsize=20)
        
        del(eff_list)
        del(eff_error_list)
        del(gamma_rate_list)

        
    # Xlabel
    ax.xaxis.set_label_coords(0.5, -0.055)
    ax.set_xlabel(r'Gamma Rate $(Hz/cm^2)$', fontsize = 22)

    # Ylabel
    ax.yaxis.set_label_coords(-0.09, 0.86)
    ax.set_ylabel(r'Efficiency (%)', fontsize = 22)
    plt.ylim([80, 100])

    # CMS format
    hfont = {'fontname':'Helvetica'}    
    plt.text(0.12, 0.89, "CMS", fontdict=hfont,  fontweight='bold', transform=plt.gcf().transFigure) # Value for on top: 0.17, 0.89, inside plot: 0.17, 0.80
    plt.text(0.22, 0.89, "Preliminary", fontdict=hfont, style='italic',fontsize = 22, transform=plt.gcf().transFigure) # Value for on top: 0.27, 0.89, inside plot: 0.27, 0.80

    ### If ones would like to move the scientific notation
    #t = ax.yaxis.get_offset_text()
    #t.set_x(0.008)

    plt.grid()
    plt.savefig("Efficiency_VS_Gamma_Rate_WP.png")
```

```python
def current_cluster_rate_WP(path, gas):
    
    surface_bot_gap = 100*70 #cm
    surface_top_gap = surface_bot_gap
    
    # Figure and axis
    fig, ax = plt.subplots()
    
    for f in gas.keys():
        leg = f
        # List to pass the density current 
        density_current_list = []
        # List to pass the error of density current
        density_current_error_list = []
        # List to pass the cluster rate
        cluster_rate_list = []
        for c in gas[f]:
            # Read the csv file as a panda dataframe
            df = pd.read_csv(path + f + '/' + c)

            # Takes the gamma cluster size    
            gamma_cs = df['gamma_CLS_WP_muon'][0]
            # Takes the hit rate
            gamma_hit_rate = df['noiseGammaRate_WP_muon'][0]
            # Compute the cluster rate
            cluster_rate = gamma_hit_rate / gamma_cs
            
            # Takes the currents
            current_top = df['imon_KODELC-TOP_WP_muon'][0]
            current_bot = df['imon_KODELC-BOT_WP_muon'][0]
           
            # Takes the total density current 
            density_current_ = current_top/surface_top_gap + current_bot/surface_bot_gap          
            density_current_list.append(density_current_)
            density_current_error_list.append(0)
            cluster_rate_list.append(cluster_rate)
        
        # Attributes xaxis and yaxis
        plt.errorbar(cluster_rate_list, density_current_list, yerr=density_current_error_list, marker='D', markersize=10,
                     linestyle='', label=leg)
        ax.legend(loc='best', fontsize=20)
        
        del(density_current_list)
        del(density_current_error_list)
        del(cluster_rate_list)

        
    # Xlabel
    ax.xaxis.set_label_coords(0.5, -0.055)
    ax.set_xlabel(r'Cluster Rate $(Hz/cm^2)$', fontsize = 22)

    # Ylabel
    ax.yaxis.set_label_coords(-0.11, 0.86)
    ax.set_ylabel(r'Density current $(\mu A/cm^2)$', fontsize = 22)
    #plt.ylim([80, 100])

    # CMS format
    hfont = {'fontname':'Helvetica'}    
    plt.text(0.12, 0.89, "CMS", fontdict=hfont,  fontweight='bold', transform=plt.gcf().transFigure) # Value for on top: 0.17, 0.89, inside plot: 0.17, 0.80
    plt.text(0.22, 0.89, "Preliminary", fontdict=hfont, style='italic',fontsize = 22, transform=plt.gcf().transFigure) # Value for on top: 0.27, 0.89, inside plot: 0.27, 0.80

    ### If ones would like to move the scientific notation
    #t = ax.yaxis.get_offset_text()
    #t.set_x(0.008)

    plt.grid()
    plt.savefig("Current_VS_Cluster_Rate_WP.png")
```

```python
def gamma_cluster_size_gamma_rate_WP(path, gas):
    
    surface_bot_gap = 0.315*1e4 #cm
    surface_top_wide_gap = 0.184*1e4 #cm
    surface_partiton_b = 0.0837*1e4 #cm
    
    # Figure and axis
    fig, ax = plt.subplots()
    
    for f in gas.keys():
        leg = f
        # List to pass the gamma cluster size
        gamma_cluster_size_list = []
        # List to pass the error of the gamma cluster size
        gamma_cluster_size_error_list = []
        # List to pass the cluster rate
        gamma_rate_list = []
        for c in gas[f]:
            # Read the csv file as a panda dataframe
            df = pd.read_csv(path + f + '/' + c)

            # Takes the gamma cluster size    
            gamma_cs = df['gamma_CLS_WP_muon'][0]
            # Takes the hit rate
            gamma_hit_rate = df['noiseGammaRate_WP_muon'][0]
            # Compute the cluster rate
            cluster_rate = gamma_hit_rate / gamma_cs
            gamma_rate = cluster_rate/df['eff_WP_muon'][0]*100
           
            gamma_cluster_size_list.append(gamma_cs)
            gamma_cluster_size_error_list.append(0)
            gamma_rate_list.append(gamma_rate)
        
        # Attributes xaxis and yaxis
        plt.errorbar(gamma_rate_list, gamma_cluster_size_list, yerr=gamma_cluster_size_error_list, marker='D', markersize=10,
                     linestyle='', label=leg)
        ax.legend(loc='best', fontsize=20)
        
        del(gamma_cluster_size_list)
        del(gamma_cluster_size_error_list)
        del(gamma_rate_list)

        
    # Xlabel
    ax.xaxis.set_label_coords(0.5, -0.055)
    ax.set_xlabel(r'Gamma Rate $(Hz/cm^2)$', fontsize = 22)

    # Ylabel
    ax.yaxis.set_label_coords(-0.09, 0.86)
    ax.set_ylabel(r'Gamma cluster size', fontsize = 22)
    plt.ylim([0, 4])

    # CMS format
    hfont = {'fontname':'Helvetica'}    
    plt.text(0.12, 0.89, "CMS", fontdict=hfont,  fontweight='bold', transform=plt.gcf().transFigure) # Value for on top: 0.17, 0.89, inside plot: 0.17, 0.80
    plt.text(0.22, 0.89, "Preliminary", fontdict=hfont, style='italic',fontsize = 22, transform=plt.gcf().transFigure) # Value for on top: 0.27, 0.89, inside plot: 0.27, 0.80
    
    ### If ones would like to move the scientific notation
    #t = ax.yaxis.get_offset_text()
    #t.set_x(0.008)

    plt.grid()
    plt.savefig("Gamma_Cluster_Size_VS_Gamma_Rate.png")
```

```python
def plot_cluster_size_hv_eff(path ,file, legend, gap):
    
    # List to pass the current for each csv
    cluster_size_list = []
    cluster_size_error_list = []
    # List to pass the cluster size for each csv
    hv_list = []
        
    for f in file:
        # Read the csv file as a panda dataframe
        df = pd.read_csv(path + f)       
            
        # Takes the cs for each gap     
        gamma_cs = df['gammaCLS']
        # Takes the cs error
        gamma_cs_error = df['gammaCLS_err']
        
        # Takes the eff voltage
        eff_volt = df['hveff_' + gap]
       
        # Insert the desired values on the lists
        cluster_size_list.append(gamma_cs)
        cluster_size_error_list.append(gamma_cs_error)
        hv_list.append(eff_volt)
        
    # Figure and axis
    fig, ax = plt.subplots()
    # Attributes xaxis and yaxis
    
    colors = np.array(['#1f77b4', '#aec7e8', '#ff7f0e', '#ffbb78', '#2ca02c',
                       '#98df8a', '#d62728', '#ff9896', '#9467bd', '#c5b0d5',
                       '#8c564b', '#c49c94', '#e377c2', '#f7b6d2', '#7f7f7f',
                       '#c7c7c7', '#bcbd22',])
    markers = np.array(["1", "8", "s", "p", "*", "+", "X", "d", "P", ">", "x", "D", "H"])
    
    # Loop used for plot the data for each csv file
    for (ef, er, vo, leg, col, ma) in zip(cluster_size_list, cluster_size_error_list, hv_list, legend, colors, markers):
        #ax.plot(vo, ef, marker='s', linestyle=None, linewidth=0)
        plt.errorbar(vo, ef, yerr=er, marker=ma, markersize=10, linestyle='', label=leg, mfc=col), #marker='.', linestyle=None, linewidth=0)
        ax.legend(loc='best', fontsize=16,)# bbox_to_anchor=(0.2, 0.2),)

    # Xlabel
    ax.xaxis.set_label_coords(0.5, -0.055)
    ax.set_xlabel(r'$HV_{eff} (V)$', fontsize = 22)

    # Ylabel
    ax.yaxis.set_label_coords(-0.09, 0.86)
    ax.set_ylabel(r'Gamma Cluster Size', fontsize = 22)
    #plt.legend('labels')

    # CMS format
    hfont = {'fontname':'Helvetica'}    
    plt.text(0.12, 0.89, "CMS", fontdict=hfont,  fontweight='bold', transform=plt.gcf().transFigure) # Value for on top: 0.17, 0.89, inside plot: 0.17, 0.80
    plt.text(0.22, 0.89, "Preliminary", fontdict=hfont, style='italic',fontsize = 22, transform=plt.gcf().transFigure) # Value for on top: 0.27, 0.89, inside plot: 0.27, 0.80

    ### If ones would like to move the scientific notation
    #t = ax.yaxis.get_offset_text()
    #t.set_x(0.008)

    plt.grid()
    
    plt.savefig('Gamma_Cluster_Size_VS_HVeff_' + gap + '.png')
```
